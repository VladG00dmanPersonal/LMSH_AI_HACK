- статус

```python
def format(file, has_region=False) -> pd.DataFrame:
    id = []
    surname = []
    first_initial = []
    second_initial = []
    class_partake = []
    class_study = []
    score = []
    title = []
    for line in file:
        line = line.split()
        
        if not line:
            continue
        try:
            id.append((int)(line[0]))
        except ValueError:
            continue
        surname.append(line[1])
        first_initial.append(line[2][:-1])
        title.append(line[-1])
        score.append((int)(line[-2]))
        if has_region:
            class_study.append((int)(line[-4]))
            class_partake.append((int)(line[-5]))
        else:
            class_study.append((int)(line[-3]))
            class_partake.append((int)(line[-4]))
        if len(line) == 7:
            second_initial.append(None)
        else:
            second_initial.append(line[3][:-1])
    
    df = pd.DataFrame({
        'id': id,
        'surname': surname,
        'first_inital': first_initial,
        'second_initial': second_initial,
        'class_partake': class_partake,
        'class_study': class_study,
        'score': score,
        'title': title
    })
    return df


files = [MOW_SCH_9, MOW_SCH_10, MOW_SCH_11, MOW_MUN_9, MOW_MUN_10, MOW_MUN_11, REG]
files_names = [
    ('MOW_SCH_9', False), 
    ('MOW_SCH_10', False), 
    ('MOW_SCH_11', False), 
    ('MOW_MUN_9', True), 
    ('MOW_MUN_10', True), 
    ('MOW_MUN_11', True), 
    ('REG', True)
]
files_names = [(files_names[i][0] + '.csv', files[i][1]) for i in range(len(files_names))]
for i in range(len(files)):
    format(files[i][0], files[i][1]).to_csv(os.path.join('./parsed', files_names[i]), index=False)
```
