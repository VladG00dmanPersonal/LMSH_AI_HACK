**DataFrame:**
- id
- Имя
- 1 инициал
- 2 инициал
- класс обучения
- класс участия
- балл
- статус

```python
def format(file) -> pd.DataFrame:
    id = []
    surname = []
    first_initial = []
    second_initial = []
    class_partake = []
    class_study = []
    score = []
    title = []
    for line in file:
        line = line.split()
        try:
            id.append((int)(line[0]))
        except ValueError:
            continue
        surname.append(line[1])
        first_initial.append(line[2][:-1])
        title.append(line[-1])
        score.append((int)(line[-2]))
        class_study.append((int)(line[-3]))
        class_partake.append((int)(line[-4]))
        if len(line) == 7:
            second_initial.append(None)
        else:
            second_initial.append(line[3][:-1])
    
    df = pd.DataFrame({
        'id': id,
        'surname': surname,
        'first_inital': first_initial,
        'second_initial': second_initial,
        'class_partake': class_partake,
        'class_study': class_study,
        'score': score,
        'title': title
    })
    return df

files = [MOW_SCH_9, MOW_SCH_10, MOW_SCH_11, MOW_MUN_9, MOW_MUN_10, MOW_MUN_11, REG]
files_names = ['MOW_SCH_9', 'MOW_SCH_10', 'MOW_SCH_11', 'MOW_MUN_9', 'MOW_MUN_10', 'MOW_MUN_11', 'REG']
for i in range(len(files)):
    format(files[i]).to_csv('/parsed/' + files_names[i] + '.csv', index=False)
```
