```python
import pandas as pd
import os
from io import TextIOWrapper
```

```python
MOW_SCH_9_2025 = open('raw/MOW_SCH_9_2025')
MOW_SCH_10_2025 = open('raw/MOW_SCH_10_2025')
MOW_SCH_11_2025 = open('raw/MOW_SCH_11_2025')
```

```python
type(MOW_SCH_9_2025)
```

```python
def format(file: TextIOWrapper, file_name) -> pd.DataFrame:
    print(f'Processing file {file_name}')

    file.seek(0)

    id = []
    surname = []
    first_initial = []
    second_initial = []
    class_study = []
    class_partake = []
    score = []
    title = []

    for line in file:
        line = line.split()
        print(line)

        id.append(line[0])
        surname.append(line[1])
        first_initial.append(line[2])
        title.append(line[-1])
        score.append((int)(line[-2]))
        class_partake.append((int)(line[-3]))
        class_study.append(line[-4])
        if len(line) == 8:
            second_initial.append(line[4].replace('.', ''))
        else:
            second_initial.append(None)

    print(f'Processed file {file_name}')

    df = pd.DataFrame({
        'id': id,
        'surname': surname,
        'first_inital': first_initial,
        'second_initial': second_initial,
        'class_study': class_study,
        'class_partake': class_partake,
        'score': score,
        'title': title
    })
    return df
```

```python
files = [
    (MOW_SCH_9_2025, 'MOW_SCH_9_2025.csv'),
    (MOW_SCH_10_2025, 'MOW_SCH_10_2025.csv'),
    (MOW_SCH_11_2025, 'MOW_SCH_11_2025.csv')
]

for file, file_name in files:
    format(file, file_name).to_csv(os.path.join('../raw_data', file_name), index=False)
```
