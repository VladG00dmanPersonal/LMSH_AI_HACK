```python
import pandas as pd
from io import TextIOWrapper
```

```python
FIN_2025 = open('./raw/FIN_2025')
```

```python
def format(file: TextIOWrapper) -> pd.DataFrame:
    file.seek(0)

    id = []
    surname = []
    name = []
    father_name = []
    class_study = []
    region = []
    score = []
    title = []

    for line in file:
        if line == '120	Ражабов Сайидазим Зариф Угли	10	г. Москва	502	призёр\n':
            id.append(120)
            surname.append('Ражабов')
            name.append('Саёидазим')
            father_name.append('Зариф Угли')
            class_study.append(10)
            region.append('г. Москва')
            score.append(502)
            title.append('призёр')
            continue
        elif line == '182	Эль Бахрави Амр Ихаб Абдель Хамид	10	Московская область	465	призёр\n':
            id.append(182)
            surname.append('Эль Бахрави')
            name.append('Амр Ихаб')
            father_name.append('Абдель Хамид')
            class_study.append(10)
            region.append('Московская область')
            score.append(465)
            title.append('призёр')
            continue

        line = line.split(' ')
        for i in range(len(line)):
            line[i] = line[i].split('\t')
        tmp = line
        line = []
        for t in tmp:
            line.extend(t)
        # print(line)
        
        id.append((int)(line[0].replace('.', '')))
        surname.append(line[1])
        name.append(line[2])

        i = 1
        try:
            a = (int)(line[3])
            father_name.append(None)
            i = 0
        except ValueError:
            father_name.append(line[3])

        class_study.append((int)(line[3 + i]))

        region.append(line[(4 + i):-2])
        a = ''
        for j in region[-1]:
            a = a + ' ' + j
        region[-1] = a.strip()

        title.append(line[-1].strip())
        score.append((int)(line[-2]))

    df = pd.DataFrame({
        'id': id,
        'surname': surname,
        'name': name,
        'father_name': father_name,
        'class_study': class_study,
        'region': region,
        'score': score,
        'title': title
    })
    return df
```

```python
format(FIN_2025).to_csv('../raw_data/FIN_2025.csv', index=False)
```
