```python
import pandas as pd
from io import TextIOWrapper
```

```python
def format(file: TextIOWrapper) -> pd.DataFrame:
    file.seek(0)

    id = []
    surname = []
    name = []
    father_name = []
    region = []
    class_study = []
    p = []
    title = []

    for line in file:
        if line == '389	Запольски Мишель (Москва, 11 класс)	100	23	7	.	100	88	0	.	318	\n':
            id.append(389)
            surname.append('Запольски')
            name.append('Мишель')
            father_name.append(None)
            region.append('Москва')
            class_study.append(11)
            p.append([100, 23, 7, 0, 100, 88, 0, 0, 318])
            title.append(None)
            continue
        elif line == '493	Разина Алиса Сергеевна (Смоленская область, 11 класс)	25	.	9	.	29	.	.	.	63':
            id.append(493)
            surname.append('Разина')
            name.append('Алиса')
            father_name.append('Сергеевна')
            region.append('Смоленская область')
            class_study.append(11)
            p.append([25, 0, 9, 0, 29, 0, 0, 0, 63])
            title.append(None)
            continue
            

        line = line.split(' ')
        for i in range(len(line)):
            line[i] = line[i].split('\t')
        tmp = line
        line = []
        for t in tmp:
            line.extend(t)
        print(line)

        id.append(line[0])
        surname.append(line[1])
        name.append(line[2])
        father_name.append(line[3])

        i = 0
        a = line[4]
        while not line[4 + i].endswith(','):
            i += 1
            a = a + ' ' + line[4 + i]
        a = a.replace('(', '')
        a = a.replace(',', '')
        region.append(a)

        class_study.append(line[5 + i])
        a = line[(i + 7) : (i + 16)]
        for j in range(9):
            if a[j] == '.':
                a[j] = 0
            a[j] = (int)(a[j])
        p.append(a)

        if 16 + i < len(line):
            title.append(line[16 + i])
        else:
            title.append(None)

        print(len(id), len(surname), len(name), len(father_name), len(region), len(class_study), len(p), len(title))

        # print(id[-1], surname[-1], name[-1], father_name[-1], region[-1], class_study[-1], *p[-1])
    
    df = pd.DataFrame({
        'id': id,
        'surname': surname,
        'name': name,
        'father_name': father_name,
        'region': region,
        'class_study': class_study,
        'p1': [i[0] for i in p],
        'p2': [i[1] for i in p],
        'p3': [i[2] for i in p],
        'p4': [i[3] for i in p],
        'p5': [i[4] for i in p],
        'p6': [i[5] for i in p],
        'p7': [i[6] for i in p],
        'p8': [i[7] for i in p],
        'score': [i[8] for i in p],
        'title': title
    })
    return df
```
