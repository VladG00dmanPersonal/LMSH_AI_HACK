```python
import pandas as pd
import os
from io import TextIOWrapper
```

```python
MOW_SCH_9 = open('./raw/MOW_SCH_9')
MOW_SCH_10 = open('./raw/MOW_SCH_10')
MOW_SCH_11 = open('./raw/MOW_SCH_11')
MOW_MUN_9 = open('./raw/MOW_MUN_9')
MOW_MUN_10 = open('./raw/MOW_MUN_10')
MOW_MUN_11 = open('./raw/MOW_MUN_11')
MOW_REG = open('./raw/MOW_REG')
MOSH = open('./raw/mosh')

MOW_MUN_9_2025 = open('raw/MOW_MUN_9_2025')
MOW_MUN_10_2025 = open('raw/MOW_MUN_10_2025')
MOW_MUN_11_2025 = open('raw/MOW_MUN_11_2025')
MOW_REG_2025 = open('raw/MOW_REG_2025')
```

**DataFrame:**
- id
- Имя
- 1 инициал
- 2 инициал
- класс обучения
- класс участия
- балл
- статус

```python
def format(file: TextIOWrapper, file_name, has_region=True, has_score=True, fill_region='77') -> pd.DataFrame:
    print(f'Creating {file_name}')

    file.seek(0)
    
    id = []
    surname = []
    first_initial = []
    second_initial = []
    class_partake = []
    class_study = []
    region = []
    score = []
    title = []

    for line in file:
        # print(line)

        line = line.split()
        if not line:
            continue

        try:
            id.append((int)(line[0]))
        except ValueError:
            continue


        # id surname i1 i2? study(cat) participate region?(fill) score? title

        surname.append(line[1])
        first_initial.append(line[2][:-1])
        if line[3].endswith('.'):
            second_initial.append(line[3][:-1])
        else:
            second_initial.append(None)
        
        k = 0
        title.append(line[-1])
        if title[-1] == 'степени':
            title[-1] = line[-3] + line[-2] + line[-1]
            k = -2
        if has_score:
            score.append((int)(line[-2 + k].replace(',', '.')))
        else:
            score.append(None)
        if has_region:
            region.append(line[-2 - has_score + k])
        else:
            region.append(fill_region)
        class_partake.append((int)(line[-2 - has_score - has_region + k]))
        class_study.append(line[-3 - has_score - has_region + k])
        
    
    df = pd.DataFrame({
        'id': id,
        'surname': surname,
        'first_inital': first_initial,
        'second_initial': second_initial,
        'class_study': class_study,
        'class_partake': class_partake,
        'region': region,
        'score': score,
        'title': title
    })

    print(f'Created {file_name}')

    return df
```

```python
files = [
    (MOW_SCH_9, 'MOW_SCH_9.csv', False, True), 
    (MOW_SCH_10, 'MOW_SCH_10.csv', False, True), 
    (MOW_SCH_11, 'MOW_SCH_11.csv', False, True), 

    (MOW_MUN_9, 'MOW_MUN_9.csv', True, True), 
    (MOW_MUN_10, 'MOW_MUN_10.csv', True, True), 
    (MOW_MUN_11, 'MOW_MUN_11.csv', True, True), 

    (MOW_REG, 'MOW_REG.csv', True, False),

    (MOSH, 'MOSH.csv', False, True),

    (MOW_MUN_9_2025, 'MOW_MUN_9_2025.csv', True, True),
    (MOW_MUN_10_2025, 'MOW_MUN_10_2025.csv', True, True),
    (MOW_MUN_11_2025, 'MOW_MUN_11_2025.csv', True, True),

    (MOW_REG_2025, 'MOW_REG_2025.csv', True, False)
]

for file, file_name, has_region, has_score in files:
    df = format(file, file_name, has_region, has_score)
    df.to_csv(os.path.join('../raw_data', file_name), index=False)
```
