```python
import numpy as np
import pandas as pd
import os
```

```python
FUTURE_9 = open('./raw/shag_v_budushee_9')
FUTURE_10 = open('./raw/shag_v_budushee_10')
FUTURE_11 = open('./raw/shag_v_budushee_11')
```

```python
def format(file) -> pd.DataFrame:

    file.seek(0)

    id = []
    surname = []
    name = []
    father_name = []
    title = []

    for line in file:
        line = line.split()

        id.append(line[0])
        surname.append(line[1])
        name.append(line[2])
        title.append(line[-1])
        if len(line) == 5:
            father_name.append(line[-2])
        else:
            father_name.append(None)
    
    df = pd.DataFrame({
        'id': id,
        'surname': surname,
        'name': name,
        'father_name': father_name,
        'title': title
    })
    
    return df
```

```python
files = [
    (FUTURE_9, 'FUTURE_9.csv'),
    (FUTURE_10, 'FUTURE_10.csv'),
    (FUTURE_11, 'FUTURE_11.csv')
]

for file, file_name in files:
    format(file).to_csv(os.path.join('../raw_data', file_name), index=False)
```
