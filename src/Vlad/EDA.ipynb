### Bit of preparing

```python
df_innopolis2425_1 = pd.read_csv(raw_data_dir + innopolis_dir + year2425_dir + innopolis_filename + "_1.csv")
df_innopolis2425_2 = pd.read_csv(raw_data_dir + innopolis_dir + year2425_dir + innopolis_filename + "_2.csv")
df_innopolis2425_3 = pd.read_csv(raw_data_dir + innopolis_dir + year2425_dir + innopolis_filename + "_3.csv")

df_innopolis2526_1 = pd.read_csv(raw_data_dir + innopolis_dir + year2526_dir + innopolis_filename + "_1.csv")
df_innopolis2526_2 = pd.read_csv(raw_data_dir + innopolis_dir + year2526_dir + innopolis_filename + "_2.csv")
df_innopolis2526_3 = pd.read_csv(raw_data_dir + innopolis_dir + year2526_dir + innopolis_filename + "_3.csv")

df_innopolis2425_1['stage'] = 1
df_innopolis2425_2['stage'] = 2
df_innopolis2425_3['stage'] = 3

df_innopolis2526_1['stage'] = 1
df_innopolis2526_2['stage'] = 2
df_innopolis2526_3['stage'] = 3

df_innopolis2425 = pd.concat([df_innopolis2425_1, df_innopolis2425_2, df_innopolis2425_3])
df_innopolis2526 = pd.concat([df_innopolis2526_1, df_innopolis2526_2, df_innopolis2526_3])

df_innopolis2425['year'] = '24-25'
df_innopolis2425['name'] = df_innopolis2425['name'].fillna("NO_NAME")
df_innopolis2425 = df_innopolis2425.sort_values(['stage', 'grade', 'total'], ascending=[True, False, False])

df_innopolis2526['year'] = '25-26'
df_innopolis2526['name'] = df_innopolis2526['name'].fillna("NO_NAME")
df_innopolis2526 = df_innopolis2526.sort_values(['stage', 'grade', 'total'], ascending=[True, False, False])

df_innopolis = pd.concat([df_innopolis2425, df_innopolis2526])

task_labmda = lambda x: 0 if x == '.' else float(x)

df_innopolis['A'] = df_innopolis['A'].fillna(0).apply(task_labmda)
df_innopolis['B'] = df_innopolis['B'].fillna(0).apply(task_labmda)
df_innopolis['C'] = df_innopolis['C'].fillna(0).apply(task_labmda)
df_innopolis['D'] = df_innopolis['D'].fillna(0).apply(task_labmda)
df_innopolis['E'] = df_innopolis['E'].fillna(0).apply(task_labmda)
df_innopolis['F'] = df_innopolis['F'].fillna(0).apply(task_labmda)

def get_total(x):
    if x['total'] == np.nan:
        return x['total']
    else:
        return x['A'] + x['B'] + x['C'] + x['D'] + x['E'] + x['F']

df_innopolis['total'] = df_innopolis.apply(get_total, axis=1)

df_innopolis['result'] = df_innopolis['result'].fillna("Deny")

df_innopolis = df_innopolis.sort_values(['year', 'stage', 'grade', 'total'], ascending=[False, True, False, False])

df_innopolis.head()
```
